```python
sta = [['a','b','c','d','e'],
       ['a','c','d','l','k'],
       ['w','e','i','r','d'],
       ['a','s','h','o','l']]

 

class parameters:
    profit = False
    low_price = False
    user_satisfaction = False

    def __init__(self,profit,low_price,user_satisfaction):
        self.profit = profit
        self.low_price = low_price
        self.user_satisfaction = user_satisfaction

    def setProfit(self, profit):
        self.profit = profit

    def setLowPrice(self, low_price):
        self.low_price = low_price

    def setUserSat(self,user_sat):
        self.user_satisfaction = user_sat

N_reg = parameters(False,True,True)
S_reg = parameters(True,True,True)
E_reg = parameters(True,False,False)
W_reg = parameters(True,True,False)

 

Region = [N_reg, S_reg, E_reg, W_reg]
for i in range(len(Region)):
    profitable = Region[i].profit and (Region[i].low_price or Region[i].user_satisfaction)
    if profitable:
        sta[i].append('n')

print(sta)
```

```python
def func(pressure,gas_constant,temperature):
    vol=(pressure*1.0)/(gas_constant*temperature)
    return vol

gas_vol=func(10000,8.314,300)
print(f"Gas volume:{gas_vol} m^3")
```

```python
def mass(pressure,gas_constant,temperature,mass):
    m=func(pressure,gas_constant,temperature)*mass
    return m
print(mass(10000,8.314,300,44),"kg")
```

```python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
print(fact(5))
```

```python
def calc_tot_depth(segments):
    if not segments:
        return 0
    else:
        curr_seg_depth = segments[0]
        remaining_seg = segments[1:]
        return curr_seg_depth + calc_tot_depth(remaining_seg)
    
print(calc_tot_depth([1,2,3]))
```

```python
def generate_squares(n):
    for i in range(1,n+1):
        yield i ** 2
for i in generate_squares(5):
    print(i)
```

```python
import datetime
def Calculate_Monthy_Production(yearlyProduce):     
    month = 1     
    for i in range (1,13):         
        x = datetime.datetime(2023, month, 1)         
        yield({"month":x.strftime("%B"),"production":yearlyProduce/12})         
        month = month + 1

for i in Calculate_Monthy_Production(12000):
    print(f"Production in {i['month']} is {i['production']}")
```

```python
class my_cls:
    def gas_vol(self,p,t):
        gc = 8.314
        x = (p*1.0)/(gc*t)
        return x

    def mass_gas(self,p,t,mol):
        return self.gas_vol(p,t)*mol

    def factorial(self,n):
        if n==0:
            return 1
        else:
            return n*self.factorial(n-1)

    def calc_tot_depth(self,segments):
        if not segments:
            return 0
        else:
            curr_seg_depth = segments[0]
            remaining_seg = segments[1:]
            return curr_seg_depth + self.calc_tot_depth(remaining_seg)

 

x = my_cls()
print(x.gas_vol(10000,300))
print(x.mass_gas(10000,300,20))
print(x.factorial(5))
print(x.calc_tot_depth([1,2,3]))
```
